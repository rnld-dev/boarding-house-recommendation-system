```python
import re, pandas as pd, time
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import NoSuchElementException, ElementNotInteractableException, TimeoutException
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.support.ui import WebDriverWait
import logging
```

```python
executable_path = r"D:\New\Training\Learn Web Scraping\chromedriver-win64\chromedriver.exe"
service = Service(executable_path)
options = webdriver.ChromeOptions()

# Menambahkan argumen untuk menonaktifkan beberapa fitur pada Chrome yang dapat mendeteksi otomatisasi
# Ini membantu menghindari mendeteksi penggunaan Selenium WebDriver oleh website
options.add_argument("--disable-blink-features=AutomationControlled")

# Menambahkan opsi eksperimental untuk menonaktifkan fitur otomatisasi yang dapat dideteksi oleh Chrome
options.add_experimental_option("excludeSwitches", ["enable-automation"])

options.add_experimental_option("useAutomationExtension", False) 
options.add_argument('--ignore-certificate-errors')
driver = webdriver.Chrome(service=service, options=options)

LINK = "https://mamikos.com/cari/upn-veteran-jawa-timur-gunung-anyar-kota-surabaya-jawa-timur-indonesia/all/bulanan/0-15000000"
driver.get(LINK)
original_window = driver.current_window_handle # menyimpan window handle yang pertama kali dibuka
time.sleep(5)
```

### HANDLE POPUP DI MULTI PAGES

```python
from IPython.display import Image, display

# Path to the image
image_path = r"D:\New\Training\Algoritma\TA\Ref Code\scrap-mamikos\assets\multipages.png"

# Display the image with a specified width and height
display(Image(filename=image_path))
```

```python
def handle_multiple_page_popup(driver):
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    handled_popup = False
    
    while True:
        if not handled_popup:
            # Handle the "Nanti Saja" button popup first
            try:
                nanti_saja_button_xpath = '//*[@id="desktopBannerWrapped"]/div/div[3]/div[1]/button[1]'
                nanti_saja_button = WebDriverWait(driver, 10).until(EC.element_to_be_clickable((By.XPATH, nanti_saja_button_xpath)))
                nanti_saja_button.click()
                print("Popup 'Nanti Saja' ditangani.")
                handled_popup = True
                continue  # Restart the loop to handle the next actions after closing the popup
            except (NoSuchElementException, TimeoutException):
                print("Tidak dapat menemukan atau mengklik tombol 'Nanti Saja' pada popup.")
                break

        try:
            # Handle other popups by clicking on the "Saya Mengerti" button
            alert_button_xpath = '//*[@id="filterKostTypeWrapper"]/div/div[1]/div[1]/div/div/div[3]/button'
            try:
                alert_button = driver.find_element(By.XPATH, alert_button_xpath)
                alert_button.click()
                print("Popup 'Saya Mengerti' ditangani.")
            except NoSuchElementException:
                # print("Tidak dapat menemukan tombol 'Saya Mengerti' pada popup.")
                pass  # Continue to check for next button anyway
        except NoSuchElementException:
            print("Tidak ada elemen popup yang diharapkan")
            pass  # Continue to check for next button anyway

        try:
            # Try to find the next_button
            next_button = WebDriverWait(driver, 20).until(EC.element_to_be_clickable((By.CLASS_NAME, 'list__content-load-link')))
            next_button.click()

            # Wait until the next page is loaded
            WebDriverWait(driver, 20).until(
                EC.presence_of_element_located((By.XPATH, '//*[@id="app"]/div/div[5]/div/div[1]/div/div/div[1]/div[1]/div[1]'))
            )

            time.sleep(3)
            print("Next page loaded successfully.")
        
        except TimeoutException:
            print("TimeoutException: Next button is not clickable or not found.")
            break
        
        except (ElementNotInteractableException, NoSuchElementException):
            print("Tidak ada elemen next button atau tidak dapat diinteraksikan")
            break
```

```python
# HANDLE POP-UP YANG ADA SETELAH MASUK KE LINK KOS (SINGLE KOS)
def handle_single_page_popup(driver):
    # Handle "Lanjut" button
    while True:
        try:
            lanjut_button = WebDriverWait(driver, 5).until(
                EC.element_to_be_clickable((By.XPATH, '//*[@id="app"]/div/section/div[4]/div/div/div[2]/button'))
            )
            lanjut_button.click()
            logging.info("Clicked 'Lanjut'")
        except (NoSuchElementException, TimeoutException):
            break

    # Handle "Saya Mengerti" button
    try:
        saya_mengerti_button = WebDriverWait(driver, 5).until(
            EC.element_to_be_clickable((By.XPATH, '//*[@id="app"]/div/section/div[4]/div/div/div[2]/button'))
        )
        saya_mengerti_button.click()
        logging.info("Clicked 'Saya Mengerti'")
    except (NoSuchElementException, TimeoutException):
        pass

    # Handle "Nanti Saja" button for notification
    try:
        nanti_saja_button = WebDriverWait(driver, 5).until(
            EC.element_to_be_clickable((By.XPATH, '//*[@id="desktopBannerWrapped"]/div/div[3]/div[1]/button[1]'))
        )
        nanti_saja_button.click()
        logging.info("Clicked 'Nanti Saja'")
    except (NoSuchElementException, TimeoutException):
        pass
```

### ARAHKAN DRIVER UNTUK MENGULIR MULTIPLE PAGE KE BAWAH SAMPAI HABIS

```python
handle_multiple_page_popup(driver)
time.sleep(10)
```

```python
tag_all_kos = driver.find_element(By.CLASS_NAME, 'row.room-list')

# dari tag_all_kos ambil semua kos dengan nama class 'room-list__card'
all_kos = tag_all_kos.find_elements(By.CLASS_NAME, 'room-list__card')
print(f"Total kos: {len(all_kos)}")
```

```python
DATA_KOST = []
```

```python
# Tentukan start_index dan end_index
start_index = 1
end_index = 15  # Mengambil 15 kos pertama

# Loop untuk setiap kos mulai dari iterasi ke-1
for i, kos in enumerate(all_kos[start_index-1:end_index], start=start_index):
    
    # Jenis Kost
    overview_kos = kos.find_element(By.CLASS_NAME, 'kost-rc__content')
    jenis_kos = overview_kos.find_elements(By.CLASS_NAME, 'rc-overview__label.bg-c-label')
    
    # Jika jenis kos lebih dari 1 maka print data yang kedua
    if len(jenis_kos) > 1:
        jenis_kos = jenis_kos[1].text
    else:
        jenis_kos = jenis_kos[0].text
    
    # Nama Kost
    nama_kos = overview_kos.find_element(By.CLASS_NAME, 'rc-info__name')
    nama_kos = nama_kos.text

    # Harga
    harga = overview_kos.find_element(By.CLASS_NAME, 'rc-price__text')
    harga = harga.text

    # Lokasi
    lokasi = overview_kos.find_element(By.CLASS_NAME, 'rc-info__location')
    lokasi = lokasi.text
    
    # Fasilitas
    fasilitas = overview_kos.find_elements(By.CLASS_NAME, 'rc-facilities')
    fasilitas_text_list = [f.text for f in fasilitas]  # Mengambil teks dari setiap elemen
    fasilitas_text = re.sub(r'[^a-zA-Z0-9\s]', '', ' '.join(fasilitas_text_list))  # Menggabungkan dan membersihkan teks

    # Klik kos
    kos.click()
    time.sleep(2)
                            
    # Pindah driver ke tab yang terbuka
    driver.switch_to.window(driver.window_handles[1])
    time.sleep(2)
    link_kos = driver.current_url

    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")

    time.sleep(2)
    
    # Tangani pop-up jika ada
    handle_single_page_popup(driver)
    
    # Ambil data luas kamar
    luas_kamar_element = WebDriverWait(driver, 20).until(
        EC.presence_of_element_located((By.XPATH, '/html/body/div[3]/div/section/section/div[2]/div[1]/section[4]/div[1]/div/div/div[1]/div/div[2]/div/p'))
    )
    luas_kamar = luas_kamar_element.text
    
    # Ambil data Keterangan Listrik
    listrik_element = WebDriverWait(driver, 20).until(
        EC.presence_of_element_located((By.XPATH, '/html/body/div[3]/div/section/section/div[2]/div[1]/section[4]/div[1]/div/div/div[2]/div/div[2]/div/p'))
    )
    listrik = listrik_element.text
    
    # Tulis ke dataframe
    DATA_KOST.append([link_kos, nama_kos, lokasi, harga, jenis_kos, fasilitas_text, listrik, luas_kamar])
    
    # Close the tab
    driver.close()
    driver.switch_to.window(original_window)
    time.sleep(3)
    
    # Cetak iterasi ke berapa
    print(f"Iterasi ke-{i} selesai")
```

## Setelah semua iterasi selesai, buat DataFrame dan simpan sebagai CSV

```python
columns = ["Link", "Nama", "Lokasi", "Harga", "Jenis", "Fasilitas", "Listrik", "Luas Kamar"]
df = pd.DataFrame(DATA_KOST, columns=columns)
```
